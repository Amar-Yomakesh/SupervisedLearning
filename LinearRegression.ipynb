```python
import numpy as np
import matplotlib.pyplot as plt
x_train = np.array([0.1,0.2,0.3])
y_train = np.array([200,300,400])

plt.scatter(x_train,y_train,marker='x',c='r')
plt.xlabel('sq feet in 1000')
plt.ylabel('price in 1000$')
plt.show()
```

```python
def compute_model(x,w,b):
    m = x.shape[0]
    f_x = np.zeros(m)
    for i in range(m):
        f_x[i] = (x[i] * w )+ b
    return f_x
    
```

```python
f_wb = compute_model(x_train,880,130)
plt.plot(x_train,f_wb,c='b',label='Prediction plot')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.show()
```
